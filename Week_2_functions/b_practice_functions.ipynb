# Answer



```python
def add_two_numbers(a_number, b_number):
    """add the two numbers
    @param a_number - first number
    @param b_number - second number
    @return returns a + b"""

    return a_number + b_number
```

```python
# Check that it works
a = 3
b = 7
my_ret = add_two_numbers(a_number=a, b_number=b)
assert my_ret == a + b
```
